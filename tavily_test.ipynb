```python
from langchain_community.tools.tavily_search import TavilySearchResults
from dotenv import load_dotenv
load_dotenv()
```

```python
# define how many results needed per query via TavilySearchResults class 
tool = TavilySearchResults(max_results=2)
results = tool.invoke({"query": 'Indian Men Cricket team Sport updates from 2022 to 2026'})
results
```

```python
for r in results:
    print(r['content'])
```
